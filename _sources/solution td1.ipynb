$$
min(a, b) = - max(-a, -b)
$$
    

#### Remarques:
1. **max(a, b)**:
   Pour déterminer le maximum entre deux nombres `a` et `b` en utilisant leur valeur absolue, on peut utiliser la formule :
   $$
   \text{max}(a, b) = \frac{a + b + |a - b|}{2}
   $$
   

2. **min(a, b)**:
   De manière similaire, pour déterminer le minimum entre `a` et `b`, vous pouvez utiliser la formule :
   $$
   \text{min}(a, b) = \frac{a + b - |a - b|}{2}
   $$

1. **Montrer que $2-3\sqrt{2} \notin \mathbb{Q}$**

Supposons par l'absurde que $2-3\sqrt{2}$ est rationnel. Alors il pourrait être écrit sous la forme:
$$2-3\sqrt{2} = \frac{p}{q}$$
où $p$ et $q$ sont des entiers sans facteurs communs et $q \neq 0$.

En réarrangeant, nous obtenons:
$$3\sqrt{2} = 2 - \frac{p}{q} = \frac{2q-p}{q}$$

En multipliant chaque côté par $q$:
$$3q\sqrt{2} = 2q-p$$

Maintenant, divisons chaque côté par 3 :
$$q\sqrt{2} = \frac{2q-p}{3}$$

Cela signifie que $\sqrt{2}$ serait rationnel (car le ratio de deux nombres entiers est rationnel). Cependant, nous savons que $\sqrt{2}\notin \mathbb{Q}$. Par conséquent, notre hypothèse de départ était fausse et $2-3\sqrt{2} \notin \mathbb{Q}$.

2. **Montrer que $1-\dfrac{1}{\sqrt{2}} \notin \mathbb{Q}$**

Supposons par l'absurde que $1-\dfrac{1}{\sqrt{2}}$ est rationnel. Alors, il pourrait être écrit sous la forme:
$$1-\dfrac{1}{\sqrt{2}} = \frac{m}{n}$$
où $m$ et $n$ sont des entiers sans facteurs communs et $n \neq 0$.

En réarrangeant, nous obtenons:
$$\dfrac{1}{\sqrt{2}} = 1 - \frac{m}{n} = \frac{n-m}{n}$$

En multipliant chaque côté par $2$, nous obtenons:
$$\sqrt{2}= 2\frac{n-m}{n} = \frac{2(n-m)}{n}$$

Encore une fois, cela impliquerait que $\sqrt{2}$ serait rationnel. Cependant, nous savons que $\sqrt{2}\notin \mathbb{Q}$. Par conséquent, notre hypothèse de départ était fausse et $1-\dfrac{1}{\sqrt{2}} \notin \mathbb{Q}$.

**1. Démontrer que si $r \in \mathbb{Q}$ et $x \notin \mathbb{Q}$ alors $r + x \notin \mathbb{Q}$ et si $r \neq 0$ alors $rx \notin \mathbb{Q}$**



(a) Supposons par l'absurde que $r + x \in \mathbb{Q}$. Alors $(r+x) - r = x \in \mathbb{Q}$, ce qui est une contradiction avec notre hypothèse que $x \notin \mathbb{Q}$.

(b) Supposons que $rx \in \mathbb{Q}$ et $r \neq 0$. Alors, $\frac{rx}{r} = x \in \mathbb{Q}$, ce qui est aussi une contradiction. 

Remarque: Si $r=0$, alors $rx=0$ est rationnel, mais nous avons exclu ce cas.

**3. En déduire que : entre deux nombres rationnels il y a toujours un nombre irrationnel**


Soit $r_1$ et $r_2$ deux rationnels distincts. 

Sans perte de généralité, supposons que $r_1 < r_2$. Considérons le nombre $\sqrt{2} \times (r_2 - r_1) + r_1$. Ce nombre est compris entre $r_1$ et $r_2$ (Pourquoi?).

D'autre part, comme nous l'avons prouvé plus haut, $\sqrt{2} \times (r_2 - r_1)$ est irrationnel. L'addition d'un nombre rationnel et d'un nombre irrationnel donne un nombre irrationnel. Donc, $\sqrt{2} \times (r_2 - r_1) + r_1$ est irrationnel.

Par conséquent, entre deux nombres rationnels, il existe toujours au moins un nombre irrationnel.

## Exercice 4

Pour prouver que $\frac{\ln(3)}{\ln(2)}$ est irrationnel, nous allons utiliser une méthode par contradiction.


Supposons que $\frac{\ln(3)}{\ln(2)}$ soit rationnel. Cela signifie qu'il peut être exprimé comme un rapport de deux nombres entiers, c'est-à-dire:
$$
\frac{\ln(3)}{\ln(2)} = \frac{p}{q}
$$
où $p$ et $q$ sont des entiers sans facteurs communs (c'est-à-dire, ils sont copremiers) et $q \neq 0$.

En réarrangeant cette équation, nous obtenons:
$$
\ln(3) = \frac{p}{q} \ln(2)
$$

En utilisant les propriétés des logarithmes, cela donne:
$$
3 = 2^{\frac{p}{q}}
$$

En élevant chaque côté à la puissance $q$, nous obtenons:
$$
3^q = 2^p
$$

Ceci est une contradiction car:

- $3^q$ est un entier impair (toute puissance de 3 est impaire).
- $2^p$ est un entier pair (toute puissance de 2 est paire).

Donc, 
$\frac{\ln(3)}{\ln(2)}$ est irrationnel.


## Exercice 5

1. **Pour max(x,y)**

Si $x \geq y$.

Alors $|x - y| = x - y$. 

Insérons cette valeur dans la formule proposée:
$$
\frac{x + y + |x - y|}{2} = \frac{x + y + x - y}{2} = x=\text{max}(x,y)
$$



Si $y \geq x$, alors $|x - y| = y - x$ et nous trouvons de la même manière que le maximum est $y$.

2. **Pour min(x,y)**

En utilisant le même raisonnement que ci-dessus:
$$
\text{min}(x,y) = \frac{x + y - |x - y|}{2}
$$
Si $x \geq y$, alors $|x - y| = x - y$ et nous trouvons que le minimum est $y$.
Si $y \geq x$, alors $|x - y| = y - x$ et le minimum est $x$.

3. **Une formule pour max(x,y,z)**

Pour trois nombres, $x$, $y$, et $z$, on peut exprimer le maximum comme:
$$
\text{max}(x,y,z) = \text{max}(\text{max}(x,y), z)
$$
En utilisant la formule précédente pour max, cela devient:
$$
\text{max}(x,y,z) = \frac{\text{max}(x,y) + z + |\text{max}(x,y) - z|}{2}=\frac{\frac{x + y + |x - y|}{2} + z + |\frac{x + y + |x - y|}{2} - z|}{2}
$$




## Exercice 6
Soit l'ensemble $A = \{u_n | n \in \mathbb{N}\}$ où 
$$
u_n = \begin{cases} 
2^n & \text{si } n \text{ est pair} \\
2^{-n} & \text{sinon}
\end{cases}
$$

**1. Pour $n$ pair**:
La suite est donnée par $u_n = 4^k$ (où $n = 2k$). Cette suite est strictement croissante et donc n'a pas de borne supérieure.

**2. Pour $n$ impair**:
La suite est donnée par $u_n = 2^{-(2k+1)}$ (où $n = 2k + 1$). Cette suite est strictement décroissante et tend vers 0. Sa borne supérieure est 0 et sa borne inférieure est également 0 puisqu'elle ne devient jamais négative.

Donc

- L'ensemble A n'a pas de borne supérieure.
- La borne inférieure de A est 0.

## Exercice 8

**1. $supA + supB$ est un majorant de $A+B$**



Soient $a \in A$ et $b \in B$. 
Par définition du sup (borne supérieure),
$$
a \leq supA \quad \text{et} \quad b \leq supB
$$
En ajoutant les deux inégalités, on obtient :
$$
a + b \leq supA + supB
$$
Comme cela est valable pour tous les éléments $(a,b) \in A \times B$, cela signifie que $supA + supB$ est bien un majorant de l'ensemble $A+B$.

---

**2. $sup(A+B) = supA + supB$**



Nous avons déjà montré que $supA + supB$ est un majorant de $A+B$, donc il est certain que :
$$
sup(A+B) \leq supA + supB
$$

Pour montrer l'égalité, nous devons montrer que pour tout $\epsilon > 0$, il existe des éléments $a \in A$ et $b \in B$ tels que :
$$
supA + supB - \epsilon < a + b
$$

Soit $\epsilon > 0$. Comme $supA$ est la borne supérieure de $A$, il existe un élément $a' \in A$ tel que :
$$
supA - \frac{\epsilon}{2} < a'
$$

De manière similaire, comme $supB$ est la borne supérieure de $B$, il existe un élément $b' \in B$ tel que :
$$
supB - \frac{\epsilon}{2} < b'
$$

En additionnant ces deux inégalités, nous obtenons :
$$
supA + supB - \epsilon < a' + b'
$$

Cela signifie que pour tout $\epsilon > 0$, il y a toujours un élément dans $A+B$ qui est à moins de $\epsilon$ de la valeur $supA + supB$. Ainsi, $supA + supB$ est la borne supérieure de $A+B$ :
$$
sup(A+B) = supA + supB
$$


## Exercice 9

**1. $A \subset B \implies supA \leq supB$**

*Vrai.* Si $A$ est un sous-ensemble de $B$, alors chaque élément de $A$ est également un élément de $B$. Par conséquent, la borne supérieure de $A$ ne peut pas être plus grande que la borne supérieure de $B$. 

---

**2. $A \subset B \implies infA \leq infB$**

*Vrai.* Si $A$ est un sous-ensemble de $B$, chaque élément de $A$ est aussi un élément de $B$. Donc, le plus petit élément (ou la borne inférieure) de $A$ ne peut pas être plus grand que celui de $B$. 

---

**3. $sup(A \cup B) = \max(supA, supB)$**

*Vrai.* Les éléments de $A \cup B$ sont ceux de $A$ ou de $B$. Le plus grand élément de $A \cup B$ sera donc le plus grand parmi les plus grands éléments de $A$ et de $B$. 

---

**4. $sup(A+B) < supA + supB$**

*Faux en général.* En fait, comme démontré précédemment, $sup(A+B) = supA + supB$. Cependant, il est vrai que $sup(A+B)$ peut être *inférieur ou égal* à $supA + supB$.

---

**5. $sup(-A) = -infA$**

*Vrai.* Pour chaque élément $a \in A$, son opposé est $-a$ et se trouve dans $-A$. Si $infA$ est le plus petit élément de $A$, alors $-infA$ est le plus grand élément négatif de $-A$. Par conséquent, $sup(-A) = -infA$.

---

**6. $supA + infB \leq sup(A+B)$**

*Vrai.* Pour tout élément $a \in A$ et tout élément $b \in B$, $a+b$ est un élément de $A+B$. Comme $supA$ est le plus grand élément de $A$ et $infB$ est le plus petit élément de $B$, alors leur somme $supA + infB$ sera toujours un élément inférieur ou égal à n'importe quel autre élément de $A+B$. Donc, $supA + infB \leq sup(A+B)$.


## Exercice 10

**1. Pour deux réels $x$ et $y$, montrer que $|x| \geq ||x+y| - |y||$**



Utilisons le fait que $|a| \geq 0$ pour tout réel $a$ et que $|a| = a$ ou $-a$ selon le signe de $a$.

(a) Si $x+y \geq y$, alors $|x+y-y| = x$, donc $|x| = x$.
(b) Si $x+y < y$, alors $|x+y-y| = -x$, donc $|x| = -x$.

Dans les deux cas, on a:
$$
|x+y| - |y| \leq |x|
$$
et 
$$
-|x+y| + |y| \geq -|x|
$$

Cela donne
$$
||x+y| - |y|| \leq |x|
$$
Ce qui est équivalent à l'inégalité demandée.

---

**2. Montrer que pour des réels $x_1, \ldots, x_n$: $|x_1 + \ldots + x_n| \leq |x_1| + \ldots + |x_n|$**

*Preuve par récurrence*:

*Base* : Pour $n=1$, l'inégalité se réduit à $|x_1| \leq |x_1|$, ce qui est évidemment vrai.

*Hypothèse de récurrence* : Supposons que l'inégalité soit vraie pour $n=k$. Alors :
$$
|x_1 + \ldots + x_k| \leq |x_1| + \ldots + |x_k|
$$

*Étape de récurrence* : Pour $n=k+1$ :
$$
|x_1 + \ldots + x_{k+1}| = |(x_1 + \ldots + x_k) + x_{k+1}| \leq |x_1 + \ldots + x_k| + |x_{k+1}| \leq |x_1| + \ldots + |x_{k+1}|
$$

L'égalité est atteinte si et seulement si tous les $x_i$ sont du même signe.

---

**3. Comparaisons pour $E(x)$, la partie entière de $x$**:

(a) Comparer $E(x+y)$ avec $E(x)+E(y)$:

En général, $E(x+y) \leq E(x) + E(y) + 1$. Si la somme des parties fractionnaires de $x$ et $y$ est inférieure à 1, alors $E(x+y) = E(x) + E(y)$. Sinon, il y a un dépassement de 1, donc $E(x+y) = E(x) + E(y) + 1$.

(b) Comparer $E(xy)$ à $E(x)E(y)$:

Si $x$ et $y$ sont des entiers, alors $E(xy) = E(x)E(y)$.

Cependant, si l'un d'eux n'est pas un entier, il n'y a pas de garantie générale. Par exemple, si $x=1.5$ et $y=2.5$, alors $xy = 3.75$ et $E(xy) = 3$, mais $E(x)E(y) = 1 \times 2 = 2$. La relation exacte dépend des valeurs spécifiques de $x$ et $y$.
